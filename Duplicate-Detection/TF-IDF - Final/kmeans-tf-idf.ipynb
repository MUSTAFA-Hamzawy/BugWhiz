## Calculate Accuracy, Precision, Recall, and F1 Score Manually

```python
# we will calculate the accuracy, precision, recall, and f1-score of the clustering
# Initialize the variables
true_positive = 0
false_positive = 0
false_negative = 0
true_negative = 0

# Load the CSV file
df = pd.read_csv(os.path.join("..\\new_dataset\\eclipse","eclipse_small_new_updated.csv" ))

# Check if the DataFrame is empty (end of file reached)
if df.empty:
    print("End of file reached for:", file_name)
    
# loop over all the rows in the dataframe
for index, row in df.iterrows():
    dic = []
    current_bug_id = row['bug_id']
    dup_id = row['dup_id']

    # check if dup_id is not []
    if dup_id != "[]":
        
        # count the true positive , false positive, false negative, and true negative
        for centroid, data in centroids.items():
            if current_bug_id == centroid:
                # split the duplciate ids based on ;
                dup_id = dup_id.split(";")
                for i in dup_id:
                    # if i is < centroid do not count it
                    if i and int(i) in data['data_points']:
                        if i and int(i) < int(centroid):
                        # true positive is when the current bug_id is in the data_points of the centroid
                            true_positive += 1
                    else:
                        # false negative is when the current bug_id is not in the data_points of the centroid
                        # acutally it's a duplicate but it's not in the cluster
                        false_negative += 1
            else:
                for i in data['data_points']:
                    if i == current_bug_id:
                        # false positive is when the current bug_id is in the data_points of the centroid
                        # A false positive is an outcome where the model incorrectly predicts the positive class.
                        false_positive += 1
                    else:
                        # true negative is when the current bug_id is not in the data_points of the centroid
                        # A true negative is an outcome where the model correctly predicts the negative class.
                        true_negative += 1


# Calculate the accuracy, precision, recall, and f1-score
accuracy = (true_positive + true_negative) / (true_positive + false_positive + false_negative + true_negative)
precision = true_positive / (true_positive + false_positive)
recall = true_positive / (true_positive + false_negative)
f1_score = 2 * (precision * recall) / (precision + recall)

print("True Positive:", true_positive)
print("False Positive:", false_positive)
print("False Negative:", false_negative)
print("True Negative:", true_negative)

# Print the accuracy, precision, recall, and f1-score
print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1_score)


# Recall is a metric that measures how often a machine learning model correctly identifies positive instances (true positives) from all the actual positive samples in the dataset.
# Precision ==> the quality of a positive prediction made by the model. 
# F1 Score ==> the harmonic mean of precision and recall.

# Accuracy (0.976): it's identifying both duplicates and non-duplicates. 
# However, accuracy alone might not be sufficient if the dataset is imbalanced (i.e., if there are significantly more non-duplicates than duplicates which is our case).

# Precision (0.483): While this precision may seem low, it's not uncommon in scenarios where false positives are more tolerable than false negatives.

# Recall (0.879): Recall measures the proportion of true duplicate instances that the model correctly identifies from all actual duplicate instances in the dataset. 
# This suggests that the model is effective at minimizing false negatives, which is often desirable in duplicate detection tasks.

# we want to reduce false positives ==> our priority.
```

## Silhouette Score

```python
from sklearn.metrics.pairwise import euclidean_distances

# Initialize lists to store data points and labels for silhouette score calculation
all_data_points = []
labels = []

# Extract data points and labels from centroids dictionary
for centroid, data in centroids.items():
    all_data_points.extend(data['data_points'])
    labels.extend([centroid] * len(data['data_points']))

# Convert lists to numpy arrays
all_data_points = np.array(all_data_points).reshape(-1, 1)
labels = np.array(labels)

# Calculate pairwise distances between data points
distances = euclidean_distances(all_data_points)

# Initialize lists to store a_i and b_i values for each data point
a_values = []
b_values = []

# Calculate a_i and b_i values for each data point
for i in range(len(all_data_points)):
    # Get the label of the current data point
    current_label = labels[i]
    
    # Calculate a_i (average distance to other data points within the same cluster)
    same_cluster_distances = distances[i][labels == current_label]
    a_i = np.mean(same_cluster_distances)
    a_values.append(a_i)
    
    # Calculate b_i (average distance to data points in other clusters)
    other_cluster_distances = distances[i][labels != current_label]
    b_i = np.mean(other_cluster_distances)
    b_values.append(b_i)

# Calculate Silhouette score for each data point
silhouette_scores = (np.array(b_values) - np.array(a_values)) / np.maximum(np.array(a_values), np.array(b_values))

# Calculate average Silhouette score
silhouette_avg = np.mean(silhouette_scores)
print("Silhouette Score:", silhouette_avg)
```
