```python
# Import necessary libraries
import os   # Import the os module, which provides functions for interacting with the operating system
import pandas as pd # Pandas for data manipulation and analysis
import utils
```

**Load Eclipse dataset**

```python
# Get the current working directory
current_dir = os.getcwd()

# Construct the relative path to the cleaned Eclipse dataset CSV file
directory_path = os.path.join('..', 'data', 'eclipse_dataset', 'cleaned_eclipse_dataset.csv')

# Load the dataset into a DataFrame using Pandas
eclipse_dataset = pd.read_csv(os.path.join(current_dir, directory_path))
```

**Split eclipse dataset into 5 chunks to be able to upload it**

```python
# Number of chunks to split the DataFrame into
num_chunks = 30

# Split the DataFrame into chunks
chunks = utils.split_dataframe(eclipse_dataset, num_chunks)

# Specify the directory where you want to save the chunks
relative_path = os.path.join('..', 'data', 'eclipse_dataset', 'splitted_eclipse_dataset')

# Get the current working directory and combine it with the relative path
save_directory = os.path.join(current_dir, relative_path)

# Create the save directory if it does not exist
if not os.path.exists(save_directory):
    os.makedirs(save_directory)

# Save each chunk as a separate CSV file in the specified directory
for i, chunk in enumerate(chunks):
    # Construct the file path for each chunk
    chunk_file_path = os.path.join(save_directory, f'eclipse_chunk_{i+1}.csv')

    # Save the chunk as a CSV file without including the DataFrame index
    chunk.to_csv(chunk_file_path, index=False)

    # Print a success message for each chunk saved
    print(f'{chunk_file_path} saved successfully')
```
