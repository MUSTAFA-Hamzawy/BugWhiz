```python
import pandas as pd
import os
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import csv
import matplotlib.pyplot as plt
```

```python
# Define the directory paths
directories = [
    ("..\\new_dataset\\eclipse", "..\\new_dataset\\eclipse"),
    #("..\\new_dataset\\eclipse_test", "..\\new_dataset\\eclipse_test"),
    #("..\\new_dataset\\firefox", "..\\new_dataset\\firefox"),
    #("..\\new_dataset\\netbeans", "..\\new_dataset\\netbeans"),
    #("..\\new_dataset\\openoffice", "..\\new_dataset\\openoffice")
]
```

```python
# Initialize TF-IDF vectorizer
tfidf_vectorizer = TfidfVectorizer()

# Concatenate preprocessed descriptions from all files in all directories
all_preprocessed_desc = []

# Iterate over each directory
for source_dir, target_dir in directories:
    # Iterate over each file in the directory
    for file_name in os.listdir(source_dir):
        print(file_name)

        # Check if the file is a CSV file        
        if file_name.endswith(".csv"):
            # Load the CSV file
            df = pd.read_csv(os.path.join(source_dir, file_name))
            
            # Check if the DataFrame is empty (end of file reached)
            if df.empty:
                print("End of file reached for:", file_name)
                continue

            # Get the data from the preprocessed_description column
            short_desc = df['preprocessed_description'].astype(str)

            # Concatenate the preprocessed descriptions
            all_preprocessed_desc.extend(short_desc)

    print("Preprocessing done for all files in:", source_dir)

# Fit TF-IDF vectorizer on combined data to build the vocabulary
tfidf_vectorizer.fit(all_preprocessed_desc)

```

```python
# Read the CSV file into a DataFrame
df = pd.read_csv('similarity_0.9.csv')

# Initialize centroids dictionary to store centroid and its assigned data points
centroids = {}

counter = 0
# Iterate over all the rows in the DataFrame
for index, row in df.iterrows():
    centroid = row['centroid']
    data_point = row['data_point']
    distance = row['distance']

    # Check if the centroid already exists in centroids dictionary
    if centroid not in centroids:
        # If not, initialize a new cluster with the centroid and assign the data_point with its distance
        centroids[centroid] = {'data_points': [data_point], 'distances': [distance]}
        counter += 1
    else:
        # If the centroid already exists, append the data_point and its distance to the cluster
        centroids[centroid]['data_points'].append(data_point)
        centroids[centroid]['distances'].append(distance)

# Print the centroids and their assigned data points
for centroid, data in centroids.items():
    print("Centroid:", centroid)
    print("Assigned Data Points:", data['data_points'])
    print("Distances:", data['distances'])
    print()

print("Number of clusters:", counter)
```

```python
from sklearn.metrics.pairwise import euclidean_distances

# Initialize lists to store data points and labels for silhouette score calculation
all_data_points = []
labels = []

# Extract data points and labels from centroids dictionary
for centroid, data in centroids.items():
    all_data_points.extend(data['data_points'])
    labels.extend([centroid] * len(data['data_points']))

# Convert lists to numpy arrays
all_data_points = np.array(all_data_points).reshape(-1, 1)
labels = np.array(labels)

# Calculate pairwise distances between data points
distances = euclidean_distances(all_data_points)

# Initialize lists to store a_i and b_i values for each data point
a_values = []
b_values = []

# Calculate a_i and b_i values for each data point
for i in range(len(all_data_points)):
    # Get the label of the current data point
    current_label = labels[i]
    
    # Calculate a_i (average distance to other data points within the same cluster)
    same_cluster_distances = distances[i][labels == current_label]
    a_i = np.mean(same_cluster_distances)
    a_values.append(a_i)
    
    # Calculate b_i (average distance to data points in other clusters)
    other_cluster_distances = distances[i][labels != current_label]
    b_i = np.mean(other_cluster_distances)
    b_values.append(b_i)

# Calculate Silhouette score for each data point
silhouette_scores = (np.array(b_values) - np.array(a_values)) / np.maximum(np.array(a_values), np.array(b_values))

# Calculate average Silhouette score
silhouette_avg = np.mean(silhouette_scores)
print("Silhouette Score:", silhouette_avg)
```

```python
# Read the CSV file into a DataFrame
df = pd.read_csv('similarity_0.9.csv')

# Extract data points and convert them to a NumPy array
data_points = df[['data_point', 'distance']].values

# Initialize K-means clustering with the desired number of clusters (k)
k = 38  # Adjust as needed
kmeans = KMeans(n_clusters=k, random_state=42)

# Fit K-means clustering to the data points
kmeans.fit(data_points)

# Get cluster labels and centroids
cluster_labels = kmeans.labels_
centroids = kmeans.cluster_centers_

# Visualize the clusters
plt.figure(figsize=(30, 6))

# Plot each data point with its cluster color
for i in range(k):
    plt.scatter(data_points[cluster_labels == i, 0], data_points[cluster_labels == i, 1], label=f'Cluster {i+1}')

# Plot centroids
plt.scatter(centroids[:, 0], centroids[:, 1], marker='x', color='black', label='Centroids')

plt.title('K-means Clustering')
plt.xlabel('Data Point')
plt.ylabel('Distance')
plt.xticks(np.arange(data_points[:, 0].min(), data_points[:, 0].max(), 6), rotation=90)  # Set x-axis ticks every 10 units and rotate them vertically
plt.grid(True)
plt.show()
```

```python
from sklearn.manifold import TSNE

# Initialize t-SNE with 2 components for 2D visualization
# tsne stands for t-distributed Stochastic Neighbor Embedding
tsne = TSNE(n_components=2, random_state=42)

# Fit and transform data points to 2D space
data_points_2d = tsne.fit_transform(data_points)

# Visualize the clusters
plt.figure(figsize=(10, 8))

# Plot each data point with its cluster color
for i in range(k):
    plt.scatter(data_points_2d[cluster_labels == i, 0], data_points_2d[cluster_labels == i, 1], label=f'Cluster {i+1}')

plt.title('t-SNE Visualization of K-means Clustering')
plt.xlabel('Component 1')
plt.ylabel('Component 2')
plt.legend()
plt.grid(True)
plt.show()
```
